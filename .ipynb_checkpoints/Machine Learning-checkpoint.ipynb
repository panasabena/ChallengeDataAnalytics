### Como el archivo tiene muchos más registros de pacientes que son promotores que detractores, vamos a crear uno que tenga la misma cantidad de cada clasificación por igual, para que no haya sesgos.

Creo un dataset con los "Passive"

```python
# Paso 1: Filtrar el DataFrame con la condición deseada
df_filtrado = data[data['NPS'] == 'Passive']

# Paso 2: Seleccionar aleatoriamente los registros
df_aleatorio = df_filtrado.sample(frac=1)

# Paso 3: Limitar la selección a los primeros 500 registros
df_passive = df_aleatorio.head(500)

# Imprimir el DataFrame resultante
df_passive.head()
```

Creo un dataset con los "Detractor"

```python
# Paso 1: Filtrar el DataFrame con la condición deseada
df_filtrado = data[data['NPS'] == 'Detractor']

# Paso 2: Seleccionar aleatoriamente los registros
df_aleatorio = df_filtrado.sample(frac=1)

# Paso 3: Limitar la selección a los primeros 500 registros
df_detractor = df_aleatorio.head(500)

# Imprimir el DataFrame resultante
df_detractor.head()
```

Creo un dataset con los "Promotor"

```python
# Paso 1: Filtrar el DataFrame con la condición deseada
df_filtrado = data[data['NPS'] == 'Promotor']

# Paso 2: Seleccionar aleatoriamente los registros
df_aleatorio = df_filtrado.sample(frac=1)

# Paso 3: Limitar la selección a los primeros 500 registros
df_promotor = df_aleatorio.head(500)

# Imprimir el DataFrame resultante
df_promotor.head()
```

## Nos estamos enfrentando a un problema de clasificación, por lo que hay que tratar de identificar de acuerdo a las variables que tenemos que NPS obtiene el paciente

Ahora concateno los 3 datasets creados para tener uno equitativo de 1500 registros

```python
# Paso 1: Concatenar los DataFrames verticalmente
df_concatenado = pd.concat([df_promotor, df_passive, df_detractor])

# Imprimir el DataFrame resultante
df_concatenado
```

Guardo el archivo equitativo en un csv:

```python
df_concatenado.to_csv('archivo_1500.csv', index=False)
```
