### 1. Árbol de Decición

Ahora que tenemos nuestras variables predictoras `X`, por un lado, y nuestra varible a predecir `y`, por el otro, vamos a crear un modelo de árbol de decisión.

Recordemos el flujo de trabajo de ML en Scikit-Learn: creamos el modelo / entrenamos / predecimos.

```python
from sklearn.tree import DecisionTreeClassifier

# Instanciamos un objeto de la clase DecisionTreeClassifier

clf = DecisionTreeClassifier(max_depth = 10, random_state = 42) 
```

Notar que establecimos max_depth en 10.

Esto es un `hiperparámetro` que define la profundidad que tendrá nuestro árbol, es decir, cuántos niveles de preguntas hará hasta que llegue a una hoja donde asigne una etiqueta.

```python
# Entrenamos el modelo

clf.fit(X.values,y.values)
```

* (clf.classes_) nos dice como están clasificadas nuestras variables
* (clf.n_classes_) nos dice la cantidad de variables a predecir que tenemos
* (clf.max_features_) nos dice la cantidad de features que le estamos dando al modelo
* (clf.feature_importances_) nos dice cuales son las features más importantes que tiene nuestro modelo, en este caso la variable más importante para predecir el target es el NPS Score, por lo que es evidente que sea la más importante. Es directamente proporcional al valor que toma NPS.

Pero como queremos ir un poco más allá en nuestro análisis, queremos ver cual es la variable más determinante en la 
clasificación de nuestros pacientes. Por lo que se decide sacarla del modelo.

Dado que el NPS Score puede estar definido de acuerdo a un parámetro de la compañía, como por ejemplo: ¿Que tan probable es que recomiende nuestro hospital? y el paciente va a contestar del 0 al 10, lo que queremos definir nosotros es que lo hizo elegir ese número

```python
print(clf.classes_)
print(clf.n_classes_)
print(clf.max_features_)
print(clf.feature_importances_)
```

### Volvemos a definir nuestro modelo:

```python
X_1 = df[['DIASINTERNADO', 'EM_DOCTOR', 'NS_NURSESATTITUDE', 'EM_NURSING', 'EM_IMMEDIATEATTENTION',
             'EM_OVERALL', 'DOC_ATTITUDE', 'NS_CALLBELLRESPONSE', 'OVS_OVERALLSTAFFATTITUDE',
             'DOC_TREATMENTEXPLAINATION', 'NS_NURSEPATIENCE', 'DOC_TREATMENTEFFECTIVENESS',
             'NS_NURSEPROACTIVENESS', 'OVS_SECURITYATTITUDE', 'OVS_OVERALLSTAFFPROMPTNESS', 'DOC_VISITS',
             'AD_STAFFATTITUDE', 'FNB_STAFFATTITUDE', 'AE_PATIENTSTATUSINFO', 'CE_ACCESSIBILITY',
             'AE_ATTENDEECARE', 'INR_ROOMEQUIPMENT', 'AD_TIME', 'INR_ROOMCLEANLINESS', 'DP_DISCHARGEQUERIES',
             'INR_ROOMPEACE', 'INR_ROOMAMBIENCE', 'FNB_DIETICIAN', 'AD_TARRIFFPACKAGESEXPLAINATION',
             'CE_CSAT', 'DP_DISCHARGEPROCESS', 'FNB_FOODDELIVERYTIME', 'DP_DISCHARGETIME', 'AE_ATTENDEEFOOD',
             'FNB_FOODQUALITY', 'CE_VALUEFORMONEY']]
# Etiqueta a predecir
y_1 = df.NPS
```

```python
# Definir los posibles valores de profundidad del árbol
max_depths = range(1, 21)

# Lista para almacenar las puntuaciones promedio de validación cruzada
mean_scores = []

# Realizar validación cruzada para cada valor de profundidad
for depth in max_depths:
    # Crear un clasificador de árbol de decisión con la profundidad actual
    clf = DecisionTreeClassifier(max_depth=depth)
    
    # Realizar validación cruzada con 5 folds
    scores = cross_val_score(clf, X_1, y_1, cv=5)
    
    # Calcular la puntuación promedio
    mean_score = np.mean(scores)
    
    # Almacenar la puntuación promedio
    mean_scores.append(mean_score)

# Encontrar la mejor profundidad con la puntuación más alta
best_depth = max_depths[np.argmax(mean_scores)]
best_score = max(mean_scores)

# Graficar la puntuación promedio en función de la profundidad
plt.plot(max_depths, mean_scores)
plt.xlabel('Profundidad del árbol')
plt.ylabel('Puntuación promedio')
plt.title('Evaluación de la profundidad del árbol')
plt.grid(True)
plt.show()

print(f"La mejor profundidad del árbol es {best_depth} con una puntuación promedio de {best_score}")
```

```python
## Volvemos a entrenar nuestro modelo
# Entrenamos el modelo

clf.fit(X_1.values,y_1.values)
```

```python
# Predecimos

y_pred_1 = clf.predict(X_1.values)
```

## Acá observamos que el modelo funciona muy bien para una profundidad de 20, tiene un 94% de efectividad.

* Esto significa que si yo predigo el 94% de las veces cuando un paciente va a ser un Detractor, puedo anticiparme a la situación y mejorar de antemano algun servicio o brindarle algo extra.



```python
from sklearn.metrics import accuracy_score

print(accuracy_score(y_1,y_pred_1))
```
